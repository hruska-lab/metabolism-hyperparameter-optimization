```python
import sys
import pandas as pd
import nglview as nv
from rdkit import Chem
from rdkit.Chem import Draw, AllChem
from project_resources.import_utils import NotebookFinder
sys.meta_path.append(NotebookFinder())
from project_resources.cytochrome_P450 import isz_csv_data_formatting, show
```

```python
isozymes = ["3A4", "RLM", "HLC"]
rel_paths = {
    "3A4_source": r"project_resources/ChEMBL_3A4.csv",
    "3A4_sep": ";",
    "3A4": r"project_resources/3A4.csv",
    "3A4_mol_indexes": r"project_resources/molecule_indexes/3A4_mol_indexes.json",

    "RLM_source": r"project_resources/AID_1508591_datatable_all.csv",
    "RLM_sep": ",",
    "RLM": r"project_resources/RLM.csv",
    "RLM_mol_indexes": r"project_resources/molecule_indexes/RLM_mol_indexes.json",

    "HLC_source": r"project_resources/AID_1508603_datatable_all.csv",
    "HLC_sep": ",",
    "HLC": r"project_resources/HLC.csv",
    "HLC_mol_indexes": r"project_resources/molecule_indexes/HLC_mol_indexes.json"
}
smiles = {}
```

```python
for isozyme in isozymes:
    isz_csv_data_formatting(rel_paths[f"{isozyme}_source"], isozyme, sep=rel_paths[f"{isozyme}_sep"])
    df = pd.read_csv(rel_paths[isozyme])
    smiles[isozyme] = df["smiles"]
    print(f"{isozyme, list(smiles[isozyme][:10])}\n")
```

```python
# mol index 14 in RLM:
Chem.MolFromSmiles("CCO.C1=CC=C2C=C(C=CC2=C1)SC3=CC4=C(C=C3)N=C(N=C4N)N")
```

```python
# after rdkit.Chem.MolStandardize.rdMolStandardize.FragmentParent:
Chem.MolFromSmiles("Nc1nc(N)c2cc(Sc3ccc4ccccc4c3)ccc2n1")
```

```python
# after simply getting the longer portion of smiles, as seperated by the period:
Chem.MolFromSmiles("C1=CC=C2C=C(C=CC2=C1)SC3=CC4=C(C=C3)N=C(N=C4N)N")
```

```python
# first 30 molecules from RLM
_3A4_mols = []
for smi in smiles["3A4"][:30]:
    mol = Chem.MolFromSmiles(smi)
    _3A4_mols.append(mol)
Draw.MolsToGridImage(_3A4_mols, subImgSize=(500, 250))
```

```python
ligand_smi = "C[C@@H]1CN(CC(=O)N2CC(C)(C)c3nnc(Cc4ccc(F)cc4F)cc32)[C@@H](CN2Cc3c(F)cccc3C2=O)CN1"  # mol_idx 2 in 3A4.csv
try:
    mol = Chem.MolFromPDBFile("project_resources/docking_resources/ligand.pdb")
except FileNotFoundError:
    # Convert SMILES to RDKit molecule
    mol = Chem.MolFromSmiles(ligand_smi)

    # Generate 3D coordinates
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem.ETKDG())

    # Save the molecule to a file
    AllChem.MolToPDBFile(mol, "project_resources/docking_resources/ligand.pdb")

show(ligand_smi)
```
