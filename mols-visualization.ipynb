```python
import sys
import pandas as pd
import nglview as nv
from rdkit import Chem
from tdc.single_pred import ADME
from rdkit.Chem import Draw, AllChem
from project_resources.import_utils import NotebookFinder
sys.meta_path.append(NotebookFinder())
from project_resources.cytochrome_P450 import show
```

```python
mol = Chem.MolFromMolFile("project_resources/docking_resources/ligand.mol")
mol = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol)
AllChem.MMFFOptimizeMolecule(mol, maxIters=200)
Chem.MolToPDBFile(mol, "project_resources/docking_resources/ligand.pdb")
smi = Chem.MolToSmiles(mol)
show(smi)
```

```python
tdc_benchmarks = ["obach", "microsome", "hepatocyte"]
tdc_datasets = {}
smiles = {}
```

```python
obach = ADME(name='Half_Life_Obach')
obach_split = obach.get_split()
tdc_datasets["obach"] = obach_split
microsome = ADME(name='Clearance_Microsome_AZ')
microsome_split = microsome.get_split()
tdc_datasets["microsome"] = microsome_split
hepatocyte = ADME(name='Clearance_Hepatocyte_AZ')
hepatocyte_split = hepatocyte.get_split()
tdc_datasets["hepatocyte"] = hepatocyte_split
```

```python
for benchmark in tdc_benchmarks:
    smiles[benchmark] = tdc_datasets[benchmark]["train"]["Drug"]
    print(benchmark, f"{list(smiles[benchmark][:10])}\n")
```

```python
# first 30 molecules from obach
obach_mols = []
for smi in smiles["obach"][:30]:
    mol = Chem.MolFromSmiles(smi)
    obach_mols.append(mol)
Draw.MolsToGridImage(obach_mols, subImgSize=(500, 250))
```

```python
for smi in smiles["obach"][:5]:
    show(smi)
```

```python
for smi in smiles["hepatocyte"][:5]:
    show(smi)
```
